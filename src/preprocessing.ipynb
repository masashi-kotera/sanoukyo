```python
import pandas as pd

df = pd.read_excel("data/Case_Setup.xlsx")
df.head(5)
```

```python
df_NACA4 = df[df['4th digit'].isnull()].reset_index(drop=True)
df_NACA5 = df[~df['4th digit'].isnull()].reset_index(drop=True)
df_NACA4 = df_NACA4.rename(columns = {'1st digit':'M', '2nd digit':'P', '3rd digit':'XX'})
df_NACA5 = df_NACA5.rename(columns = {'1st digit':'L', '2nd digit':'P', '3rd digit':'Q','4th digit':'XX'})
df_NACA4.to_csv('data/df_NACA4.csv',index = False)
df_NACA5.to_csv('data/df_NACA5.csv',index = False)
```
